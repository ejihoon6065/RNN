```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
# 세가지
# 케라스는 텐서객체를 반드시 줄 필요 없음. 시리즈 등등 가능
# 내부적으로 tensor로 바꾸는 기능이 있음
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
```

```python
df = pd.read_csv('deeplearning_class-master/dataset/sonar.csv', header = None)
#df.columns = ['Number', 'Plasma','Diastolic','Triceps','insulin','mass','pedigree','Age','Class']
df


# 1. Number of times pregnant
# 2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
# 3. Diastolic blood pressure (mm Hg)
# 4. Triceps skin fold thickness (mm)
# 5. 2-Hour serum insulin (mu U/ml)
# 6. Body mass index (weight in kg/(height in m)^2)
# 7. Diabetes pedigree function
# 8. Age (years)
# 9. Class variable (0 or 1)
```

```python
df.info()
```

```python
dataset = df.values
dataset

df[60].unique()
```

```python
X = dataset[:,0:60].astype(float)
X
```

```python
y_obj = dataset[:,60]
y_obj
```

```python
e = LabelEncoder()
e.fit(y_obj)
Y=e.transform(y_obj)
Y

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
# Dense 개체 생성
# units 는 출력 개수, 혹은 라벨의 개수
# input_dim은 feature의 개수



IO = Dense(units=1, input_dim=60)

# 리스트로 주어야 함
# sequential 순차적 구조로 Dense를 쌓아 올림 => 여러개 쌓아올릴 경우 딥러닝
model = Sequential( [IO] )
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# loss는 cost랑 동일
# Adam의 러닝메이트 = 0.1 -> 러닝메이트 스스로 조절 가능
# optimizer의 종류는 매우 다양함
model.compile( loss = 'mean_squared_error', optimizer = Adam(0.1), metrics=['accuracy'] )



#epochs = 학습 횟수
h = model.fit(X, Y, epochs = 130, batch_size=10)
model.evaluate(x_test, y_test)
```
