```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
# 세가지
# 케라스는 텐서객체를 반드시 줄 필요 없음. 시리즈 등등 가능
# 내부적으로 tensor로 바꾸는 기능이 있음
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam


seed=0
np.random.seed(seed)
tf.random.set_seed(3)
```

```python
df1 = pd.read_csv('deeplearning_class-master/dataset/housing.csv',delim_whitespace=True ,header = None)
#df.columns = ['Number', 'Plasma','Diastolic','Triceps','insulin','mass','pedigree','Age','Class']
df1

# 1. Number of times pregnant
# 2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
# 3. Diastolic blood pressure (mm Hg)
# 4. Triceps skin fold thickness (mm)
# 5. 2-Hour serum insulin (mu U/ml)
# 6. Body mass index (weight in kg/(height in m)^2)
# 7. Diabetes pedigree function
# 8. Age (years)
# 9. Class variable (0 or 1)
dataset = df1.values
```

```python
x = dataset[:,0:13]
y = dataset[:,13]
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state=seed)
```

```python
# 딥러닝 구조를 결정합니다(모델을 설정하고 실행하는 부분입니다).
model = Sequential()


model.add(Dense(30, input_dim=13, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))
```

```python
# 딥러닝을 실행합니다.
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=200, batch_size=10)
```

```python
y_prediction = model.predict(x_test).flatten()
```

```python
for i in range(10):
    label = y_test[i]
    prediction = y_prediction[i]
    print(label, prediction)
```

```python
model.evaluate(x_test,y_test)
```
