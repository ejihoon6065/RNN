```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
# 세가지
# 케라스는 텐서객체를 반드시 줄 필요 없음. 시리즈 등등 가능
# 내부적으로 tensor로 바꾸는 기능이 있음
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
import numpy as np

seed=0
np.random.seed(seed)
tf.random.set_seed(3)
```

```python
df1 = pd.read_csv('deeplearning_class-master/dataset/wine.csv', header = None)
#df.columns = ['Number', 'Plasma','Diastolic','Triceps','insulin','mass','pedigree','Age','Class']
df1 = df1.sample(frac=1)


# 1. Number of times pregnant
# 2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
# 3. Diastolic blood pressure (mm Hg)
# 4. Triceps skin fold thickness (mm)
# 5. 2-Hour serum insulin (mu U/ml)
# 6. Body mass index (weight in kg/(height in m)^2)
# 7. Diabetes pedigree function
# 8. Age (years)
# 9. Class variable (0 or 1)
```

```python
df1.info()
```

```python
df = df1.values
df
```

```python
X = df[:,:12]
X
```

```python
Y = df[:,12]
Y
```

```python


# 리스트로 주어야 함
# sequential 순차적 구조로 Dense를 쌓아 올림 => 여러개 쌓아올릴 경우 딥러닝
model = Sequential()
model.add(Dense(30, input_dim=12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(18, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# loss는 cost랑 동일
# Adam의 러닝메이트 = 0.1 -> 러닝메이트 스스로 조절 가능
# optimizer의 종류는 매우 다양함
model.compile( loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'] )



#epochs = 학습 횟수
h = model.fit(X, Y, epochs = 200, batch_size=200)
model.evaluate(X, Y)
```
